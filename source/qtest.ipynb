```python
import numpy as np
from jabba import fastQABBA
import matplotlib.pyplot as plt
from jabba.storage import compute_storage1, compute_storage2
```

```python
num = 100
mu, sigma = 0, 1
tol, alpha = 0.01, 0.05

ts = np.random.normal(mu, sigma, num)
plt.figure(figsize=(22,5))
plt.plot(ts[:100])
plt.show()
```

```python
qabba = fastQABBA(tol=0.1, k=30, init='agg', alpha=0.1, bits_for_len=8, bits_for_inc=16)
symbols = qabba.transform(ts)
```

```python
qabba = fastQABBA(tol=0.1, k=20, init='kmeans',  bits_for_len=8, bits_for_inc=16)
symbols = qabba.transform(ts)
```

```python
qabba = fastQABBA(tol=0.1, k=30, bits_for_len=8, bits_for_inc=16)
pieces = qabba._compress(ts)
```

```python
symbols = qabba._digitize(pieces)
symbols
```

```python
size = compute_storage2(qabba.parameters.centers, len(symbols), 1)
```

```python
size/(len(ts)*32)
```

```python
inv_ts = qabba.inverse_transform(symbols, ts[0])
```

```python
plt.figure(figsize=(22,5))
plt.plot(ts[:100])
plt.plot(inv_ts[:100])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from fABBA import fABBA

ts = [np.sin(0.05*i) for i in range(1000)]  # original time series
fabba = fABBA(tol=0.1, alpha=0.1, sorting='lexi', scl=1, verbose=0)

string = fabba.fit_transform(ts)            # string representation of the time series
print(string)                               # prints aBbCbCbCbCbCbCbCA

inverse_ts = fabba.inverse_transform(string, ts[0]) # numerical time series reconstruction
```
